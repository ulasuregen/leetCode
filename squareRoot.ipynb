Let us investigate the series

$$
    x_{n+1} = \frac{1}{2} * (x_n + k / x_n)
$$

First observe that $x_n$'s are bounded from blow by $0$, since $0 \leq k$

So if one can prove that $(x_n)_{n\in\mathbb{N}}$ is decreasing. 

Then we know that $\lim_{n\to\infty} x_n$ exists.

------
# Proving Decreasing 

Let $f(x) = x - \frac{1}{2} (x + \frac{k}{x})$

Now take the derivative of $f(x) = x - \frac{1}{2}(x-\frac{k}{x} )$

$$
\begin{aligned}
    f'(x) &= \frac{1}{2}(1-\frac{k}{x^2}) \\ \\
    &= \frac{1}{2}(\frac{x^2-k}{x^2}) \\ \\
    &= \frac{1}{2x^2}(\sqrt{x} - \sqrt{k})(\sqrt{x} + \sqrt{k})
\end{aligned}
$$

If $(\sqrt{x} - \sqrt{k} \geq 0) \iff (\sqrt{x} \geq \sqrt{k})$ our function $f(x)$ is non-decreasing.


We know that Arithmetic mean always greater or equal to geometric mean:

$$
\begin{aligned}
    (\sqrt{a}-\sqrt{b})^2 &\geq 0 \\
    a - 2\sqrt{ab} + b &\geq 0 \\
    a + b &\geq 2\sqrt{ab} \\
    \frac{a+b}{2} &\geq \sqrt{ab}
\end{aligned}
$$

Now observe that 

Let 

$a = x_n$ and $b = \frac{k}{x_n}$

Then we get 

$$
x_{n+1} = \frac{1}{2} * (x_n + k / x_n) \geq \sqrt{k} 
$$ 

Thus our function $f(x)$ has to be increasing. 
Also we need to be sure that $x_0 \geq \sqrt{k}$

If $0 < k < 1 \rarr k < \sqrt{k}$

Thus choosing $x_0 = k$ won't save us. But we can cleverly choose $x_0 = k+1$

---------

Now we know $f(x) = x - \frac{1}{2} (x + \frac{k}{x})$ is increasing for $x > \sqrt k$

Observe that $\text{if } x_i = \sqrt{k} \rarr \forall(n \geq i) x_n = \sqrt{k}$ Thus we don't need to prove anything if $x_i = \sqrt{k}$ since the series directly converge to $\sqrt{k}$. Thus we can assume now on $x_{i} \neq \sqrt{k}$

Observe that $f(\sqrt{k}) = \frac{1}{2}(\sqrt{k} - \sqrt{k}) = 0$

Thus for any value $x > \sqrt{k}$ by mean value theorem we have

$$
f(x) = f(x) - f(\sqrt{k}) = |x - \sqrt{k}| * f'(c) > 0 \text{ for some } c \in (\sqrt{k}, x)
$$

Thus we get if $x_n > \sqrt{k}$ then 

$$
\begin{aligned}
f(x_n) &= x_n - \frac{1}{2} (x_n + \frac{k}{x_n}) \\
&= x_n - x_{n+1} > 0 \\
&x_n > x_{n+1}
\end{aligned}
$$

Also dont forget we know $x_n \geq \sqrt{k} \quad \forall n \in \mathbb{N}$

------

Now we know with starting point $x_0 = k + 1$, our sequence is decreasing and bounded from blow. Thus $\lim_{n\to\infty} x_n$ exists. 

And since $x_n \geq \sqrt{k}$ the $x_n$ in the dominator does not create any problem when taking the limit !!! 

$$
\begin{aligned}
\lim_{n\to\infty} x_{n+1} &= \lim_{n\to\infty} \frac{1}{2} * (x_n + k / x_n) \\
x &= \frac{1}{2} * (x + k / x) \\
2x &= x + \frac{k}{x} \\
x &= \frac{k}{x} \\
x^2 &= k \\
x &= \sqrt{k}
\end{aligned}
$$

Thus we proved if $x_0 = k+1$ our sequence converges to $\sqrt{k}$ from above

# Application

```python
import numpy as np
import pandas as pd
from plotnine import *
import math
```

```python
our_secquence = []

for i in range(4):
    n = np.random.randint(3, 6000)
    x = n + 1 # x_0 

    
    for _ in range(30):
        our_secquence.append({'k':n, 'iter':_, 'x':x, 'err': x - math.sqrt(n)})
        x = 0.5 * (x + n/x)


df = pd.DataFrame(our_secquence)
```

```python
(
    ggplot(df, aes('iter', 'err')) +
    geom_point() + 
    facet_wrap('~k')
)
```
